```python
"""
ไม้ท่อนหนึ่งยาว 20 เมตร ส่วนที่อยู่ในโครน 5 เมตร
และส่วนที่อยู่ในน้ำยาว 3.50 เมตร ส่วนที่พ้นน้ำจะยาวกี่เมตร
"""
P = float(input("ไม้ท่อนหนึ่งยาว : "))
O = float(input("ส่วนที่อยู่ในโครน : "))
N = float(input("ส่วนที่อยู่ในน้ำ   : "))
print("ส่วนที่พ้นน้ำจะต้องยาว ",(P-(O+N)),"เมตร")
```

```python
"""
นักศึกษาสาขาหนึ่งมีจำนวน 180 คน ไม่มาเรียนจำนวน  0.10 ของนักศึกษาทั้งหมด จงหานักศึกษาที่ไม่มาเรียน
"""
count = int(input("จำนวนนักศึกษาทั้งหมด : "))
absent = input("ไม่มาเรียน : ")
m       = round(int(count) *float(absent))
print("นักศึกษาที่ไม่มาเรียน",m,"คน")
```

```python
"""
ร้านขายของแห่งหนึ่ง ขายเกลือ ราคาถุงละ 18 บาท แต่ซื้อได้ไม่เกินครอบครัวล่ะ 20 ถุง
โดยติดราคา ตั้งแต่ถุงที่ 1 ถึงถุงที่ 20 ไว้ดังนี้ ให้คำนวณราคาตั้งแต่ถุงที่1-20 ว่ามีราคาเท่าไหร่บ้าง
"""
Price = int(input("ราคเกลือถุงละ = "))
print("ราคา x จำนวนถุง ")
for i in range(1,10):
    print(Price,"*" , i,"=",Price*i, "บาท")
```

```python
"""
มีน้ำอัดลม 5 ขวด และมีขนมอีก 10 ห่อ เอาไปแบ่งให้เพื่อนในกลุ่มกิน 3 คน คนละเท่าๆกันเพื่อนจะได้รับน้ำแหละขนมคนล่ะเท่าไร 
"""
a=int(input("มีน้ำอัดลม")) 
b=int(input("มีขนม")) 
c=int(input("เอาไปแบ่งให้เพื่อน")) 
print("มีน้ำอัดลมและขนมทั้งหมด",a+b,"ขวด","ห่อ") 
print("เอาไปแบ่งเพื่อน",(a+b)/c,"ลูก","ห่อ")
```

```python
"""
ซื้อแหวนเพชรราคา 32,000.0 บาท ขายไปและขาดทุน 0.050 ของราคาสร้อย  อยากทราบว่าราคาที่ขายขาดทุนเท่าไร 
"""
Price = float(input("แหวนเพชร : "))
Sale  = float(input("ขาดทุน : "))
print("ราคาที่ขาดทุน",int(Price*Sale),"บาท")
s = float(input("ราคาที่ขาดทุน : "))
print("ราคาที่ขายได้",int(Price-s),"บาท")
```

```python
"""
ในการเทคอนกรีตพื้นห้องกว้าง 5 เมตร ยาว 15 เมตร เสียค่าใช้จ่ายตารางเมตรละ 250 บาท
อยากทราบว่าจะต้องเสียเงินในการเทคอนกรีตทั้งหมดกี่บาท
"""
wide  = float(input("พื้นห้องกว้าง = "))
long  = float(input("พื้นห้องยาว  = "))
po   = int(input("จ่ายตารางเมตรละ = "))
mix   = wide * long
print("พื้นที่ห้องมี",mix, "ตารางเมตร")
pay = mix * po 
print("ต้องเสียเงินในการเทคอนกรีต",pay,"บาท")
```

```python
"""
ถ้าไปเที่ยวยุโรปต้องใช้เงินสกุล EUR โดยด้วยแลกเงินแค่ 25,000.0 บาท ต้องแลกได้เท่าไร
*กำหนด 1 EUR = 33.9329 บาท
"""
Thi = float(input("จำนวนเงินไทยที่ต้องการแลก : "))
EUR = float(input(" ค่า 1 EUR :  "))
print("จะแลกได้",((Thi*1)//EUR),"EUR")
```

```python
"""
ขายยางพาราไป 980.50 กิโลกรัม ราคากิโลกรัมละ 90.50 บาท เอาเงินที่ขายได้ไปซื้อไอแพทให้ลูก
ราคาาเครื่องล่ะ 18900 บาท จะซื้อไอแพดทั้งหมดกี่เครื่อง 
"""
V = float(input("ขายขายพาราไป : "))
S = float(input("ราคากิโลกรัมละ : "))
P = float(input("ไอแพทเครื่องล่ะ : "))
money = V * S
count = int(money // P)
print(" จะได้ไอแพด ",count, "เครื่อง")
if count < 5 :
    print("สามารถซื้อได้ แต่ได้ไม่ครบตามจำนวน")
elif count > 25 :
    print("เกินจำนวนที่ต้องการ")
else:
    count == 19
    print("ได้ครบตามจำนวน")
```

```python
"""
ไม้แผ่นหนึ่งยาว 12.00 เมตร และไม้อีกท่อนหนึ่งยาว 8.0 เมตร มาเชื่อมกันแล้วตัดออกไปทำเตียง 2.0 เมตร จะเหลืออีกกี่เมตร
"""
wood_1 = float(input("ไม้ท่อแผ่นที่ 1 ยาว : "))
wood_2 = float(input("ไม้แผ่นที่ 2 ยาว : "))
cut     = float(input("ตัดไปทำเตียง 1 เตียง ยาว : "))
Mix     = wood_1 + wood_2
wood   = Mix - cut
print("จะเหลือไม้อีก = ",wood,"เมตร")
if wood > 10 :
    print(" Use wood")
```

```python
"""
โปรโมชั่นทางร้าน ซื้ออาหารทานมากกว่า 1500 บาท ขึ้นไป ลดราคา 10 เปอร์เซ็น
------ ร้านอาหาร ------
คิดเงินค่าอาหารให้ลูกค้าโต๊ะที่ 1 รวมเงิน 1200 บาท ต้องชำระเงินกี่บาท
คิดเงินค่าอาหารให้ลูกค้าโต๊ะที่ 2 รวมเงิน 1,800 บาท ต้องชำระเงินกี่บาท
"""
while(True):
    Number = int(input("ยอดเงินรวม = "))
    if(Number<1500):
        print("ต้องชำระเงินจำนวน = ",Number,"บาท")
    else:
        Number > 1500
        sale = round(Number * 0.10)
        pay  = Number - sale
        print("คิดราคาโปรโมชั่นต้องชำระเงินจำนวน = ",pay,"บาท")
        break
```
